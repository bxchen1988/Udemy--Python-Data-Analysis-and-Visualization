```python
import numpy as np
import pandas as pd
from numpy.random import randn

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot(data =[data1,data2])
```

```python
# include outlier in wisks
sns.boxplot(data = [data1,data2],whis=np.inf)
```

```python
# horizontal boxplot
sns.boxplot(data = data1,orient= 'h')
```

```python
# normal distrn
data1 = stats.norm(0,5).rvs(100)

#two gamma distrn concatenate
data2 =np.concatenate([stats.gamma(5).rvs(50)-1,
                      -1*stats.gamma(5).rvs(50)])

# box plot both data1 and data2
sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
# violin plot = kernel density plot + box plot
# needs the kernel density plot to show the difference
sns.violinplot(data=[data1,data2])
```

```python
# control bandwith
sns.violinplot(data2,bw=0.01)
```

```python
sns.violinplot(data1,inner= 'stick', orient = 'v')
```
